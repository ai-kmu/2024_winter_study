## 4.1 벡터

### 4.1.1 벡터란

벡터 : 몇 가지 숫자를 나란히 나타낸 것으로 볼드체 소문자로 표시한다.

세로벡터 : 몇 가지 숫자를 세로로 늘어놓은 것 $a = \begin{bmatrix}1\\2\\ \end{bmatrix}$, $b = \begin{bmatrix}2\\1\\ \end{bmatrix}$

가로벡터 : 몇 가지 숫자를 가로로 늘어놓은 것 $c = \begin{bmatrix}1&2\\ \end{bmatrix}$, $d = \begin{bmatrix}1&3&5&4\\ \end{bmatrix}$

요소 : 벡터를 구성하는 숫자 하나하나를 말하는 말

벡터의 차원 : 벡터가 가지는 요소의 수 ex) a : 2차원 세로벡터, d : 4차원 가로벡터

스칼라 : 일반적인 숫자의 묶음(집합)이며 _a_,_b_와 같이 소문자 이탤릭테로 표시한다.

전치 : $T$라는 기호로 쓰이며 세로벡터를 가로벡터로, 가로벡터를 세로벡터로 변환한다.

ex) $a^\top =  \begin{bmatrix}1\\2\\ \end{bmatrix}^\top = \begin{bmatrix}1&2\\ \end{bmatrix}, d^\top = \begin{bmatrix}1&3&5&4\\ \end{bmatrix}^\top = \begin{bmatrix}1\\3\\5\\4\\ \end{bmatrix}$

### 4.1.7 내적

벡터는 내적이라고 부르는 곱셈 연산이 있으며 같은 차원을 가진 두 벡터 간의 연산에서 '$\cdot$'로 나타냅니다. 이는 대응하는 요소들을 곱한 뒤 더한 값을 취합니다.

$b \cdot c = \begin{bmatrix}1\\3\\ \end{bmatrix} \cdot \begin{bmatrix}4\\2\\ \end{bmatrix} = 1\times4 + 3\times2 = 10$

또한 변수명1.dot(변수명2)을 통해 내적을 계산합니다.

```python
b = np.array([1, 3])
c = np.array([4, 2])
print(b.dot(c))
```

내적은 b를 c로 투영한 벡터를 $b'$로 하여 $b'$와 c의 길이를 곱해 합한 것입니다.
또한 내적은 두 벡터의 유사성과 관련이 있습니다.

두 벡터가 비슷한 방향을 향할 때 내적은 큰 값을 가지고, 수직에 가까울 때, 내적은 작은 값을 갖습니다. 완전한 수직이 되면 내적은 0이 됩니다.

이를 정리하여 벡터의 내적을 수식, 도형으로 나타내보겠습니다.

image.png

마지막으로 내적은 벡터의 크기 자체에도 관계됩니다. 두 벡터의 방향이 같더라도, 한 쪽의 벡터크기가 2배가 되면 내적도 2배가 됩니다.

### 4.1.8 벡터의 크기

벡터의 크기는 |와 |의 사이에 나타냅니다 이를 통해 2, 3, D차원 벡터의 크기는 이와 같이 계산할 수 있습니다.

2차원 벡터

$|a| = |\begin{bmatrix}a_0\\a_1\\ \end{bmatrix}| = \sqrt{a_0^2+a_1^2}$

3차원 벡터

$|a| = |\begin{bmatrix}a_0\\a_1\\a_2\\ \end{bmatrix}| = \sqrt{a_0^2+a_1^2+a_2^2}$

D차원 벡터

$|a| = |\begin{bmatrix}a_0\\a_1\\...\\a_{D-1}\\ \end{bmatrix}| = \sqrt{a_0^2+a_1^2+...+a_{D-1}^2}$

코드로는 'np.linalg.norm()'으로 벡터의 크기를 구할 수 있다.

```python
a = np.array([1, 3])
print(np.linalg.norm(a))
```

### 4.4.3 중첩된 함수의 미분

중첩된 함수는 안에 있는 식을 대입하고 전개하여 미분을 계산한다

예를 들어

$f(w) = g(w)^2$

$g(w) = aw+b$

일 경우 g(w)의 식을 f(w)안의 g(w)식에 대입하고 전개하여 미분을 계산하는 방식이다.

$f(w) = (aw+b)^2 = a^2+w^2 + 2abw + b^2$

$\frac{d}{dw}f(w) = 2a^2w+2ab$

### 4.4.4 중첩된 함수의 미분 : 연쇄 법칙

하지만 식이 복잡하고 전개하기 힘든 경우에는 연쇄 법칙을 이용하여 문제를 해결합니다.

[연쇄 법칙 공식]

$\frac{df}{dw} = \frac{df}{dg} \cdot \frac{dg}{dw}$

우선 f를 g로 미분하고 나온 결과와 g를 w로 미분한 결과를 곱하여 해결한다.

위의 예시를 통해 설명하면

$\frac{df}{dg} = \frac{d}{dg}g^2 = 2g$

$\frac{dg}{dw} = \frac{d}{dw}(aw+b) = a$

이 두 수를 대입하면

$\frac{df}{dw} = \frac{df}{dg} \cdot \frac{dg}{dw} = 2(aw+b)a = 2a^2w + 2ab$

가 됩니다.

또한 이런 연쇄법칙은 3, 4중으로 중첩 가능하다. 예를들어 이러한 식이 있을 때 아래와 같은 공식을 사용합니다.

$f(w) = f(g(h(w)))$

=> $\frac{df}{dw} = \frac{df}{dg} \cdot \frac{dg}{dh} \cdot \frac{dh}{dw}$

## 4.5 편미분

### 4.5.1 편미분이란

편미분 : 복수의 변수를 갖는 함수에서 한 변수에 주목하여 다른 변수를 상수로 가정하여 미분하는 것입니다. 머신러닝에서 실제로 순수미분보다 편미분을 자주 사용합니다.

image.png

### 4.5.2 편미분과 도형

$f(w_0, w_1)$함수는 3차원 등고선 플롯으로 나타내는 그래프입니다.

image.png

즉, 편미분에 의해 임의의$(w_0, w_1)$에서의 두 방향의 기울기를 계산 가능하고 이 두 기울기를 경사(기울기 벡터, gradient)라고 부릅니다. 경사는 기울기가 가장 큰 방향과 그 크기를 나타냅니다.

image.png

### 4.5.3 경사를 그림으로 나타내기

왼쪽 그림은 f를 등고선으로 표기하고 오른쪽 그림은 w의 공간을 격자형으로 나눴을 때 각 점의 경사$\nabla_wf$를 화살표로 나타낸다

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1):
  return w0**2 + 2 * w0 * w1 + 3
def df_dw0(w0, w1):
  return 2 * w0 + 2 * w1
def df_dw1(w0, w1):
  return 2 * w0 + 0 * w1

w_range = 2
dw = 0.25
w0 = np.arange(-w_range, w_range + dw, dw)
w1 = np.arange(-w_range, w_range + dw, dw)
wn = w0.shape[0]
ww0, ww1 = np.meshgrid(w0, w1)
ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros((len(w0), len(w1)))
dff_dw1 = np.zeros((len(w0), len(w1)))
for i0 in range(wn):
  for i1 in range(wn):
    ff[i1, i0] = f(w0[i0], w1[i1])
    dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
    dff_dw1[i1, i0] = df_dw1(w0[i0], w1[i1])

plt.figure(figsize=(9, 4))
plt.subplots_adjust(wspace=0.3)
plt.subplot(1, 2, 1)
cont = plt.contour(ww0, ww1, ff, 10, colors='k')
cont.clabel(fmt='%2.0f', fontsize=8)
plt.xticks(range(-w_range, w_range+1, 1))
plt.yticks(range(-w_range, w_range+1, 1))
plt.xlim(-w_range - 0.5, w_range + .5)
plt.ylim(-w_range - 0.5, w_range + .5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

plt.subplot(1, 2, 2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)
plt.xticks(range(-w_range, w_range+1, 1))
plt.yticks(range(-w_range, w_range+1, 1))
plt.xlim(-w_range - 0.5, w_range + .5)
plt.ylim(-w_range - 0.5, w_range + .5)
plt.show()
```

결과로 보면 f는 오른쪽 위와 왼쪽 아래가 높고 왼쪽 위와 오른쪽 아래가 낮은 지형임을 볼 수 있습니다.

화살표는 각 점에서 경사가 높은 쪽을 나타내고 화살표의 크기는 경사의 가파름 정도가 길 때 화살표가 길어집니다.

화살표의 어느 지점에서 시작하더라도 그래프의 보다 높은 부분으로 진행합니다.

이렇듯 경사는 그 함수의 최대점과 최소점을 찾는 데 중요한 개념이므로, 오차함수의 최소점을 구하기 위해 오차함수의 경사를 계산합니다.

### 4.7.5 시그모이드 함수

시그모이드 함수 : $y=\frac{1}{1+e^{-1}}$으로 정의되는 매끄러운 계단같은 함수이다.

$e^{-1}$는 exp(-x)로도 나타낼 수 있다.

```python
x = np.linspace(-10, 10, 100)#-10, 10사이의 값을 100개 샘플링
y = 1/(1+np.exp(-x))

plt.figure(figsize=(4, 4))
plt.plot(x, y, 'black', linewidth=3)

plt.ylim(-1, 2)
plt.xlim(-10, 10)
plt.grid(True)
plt.show()
```

image.png

시그모이드 함수는 음에서 양의 실수를 0에서 1까지의 사이로 변환하기 때문에 확률을 나타낼 때 자주 사용합니다.

시그모이드의 미분이 자주 나오므로 설명하자면, $f(x) = 1+exp(-x)$라고 하면

$(\frac{1}{f(x)}) = \frac{f(x)'}{f(x)^2}$

이때 f(x)의 미분은 f'(x) = -exp(-x)이므로 아래의 내용을 얻습니다

$y' = (\frac{1}{1+exp(-x)})' = -\frac{-exp(-x)}{(1+exp(-x))^2} = \frac{exp(-x)}{(1+exp(-x))^2}$

이 식을 조금 변형하면

$y' = \frac{1}{1+exp(-x)} ⋅ -\frac{1+exp(-x)-1}{1+exp(-x)}$

$y' = \frac{1}{1+exp(-x)} ⋅ (1 - \frac{1}{1+exp(-x)})$

여기서 1/(1+exp(-x))는 y자체였기 때문에 y로 바꿔쓰면 깔끔한 형태가 됨

$y' = y(1-y)$

### 4.7.6 소프트맥스 함수

세 개의 수$x_0=2, x_1=1, x_2=-1$이 있고 이 수들의 대소 관계를 유지하는 확률 $y_0, y_1, y_2$로 나타내려고 한다 각 확률은 0~1사이의 숫자가 나오며 합치면 1이 된다.

이때 각 $x_i$의 exp의 합계 u를 구합니다.

$u = exp(x_0) + exp(x_1) + exp(x_2)$

이를 통해 변환한 y값들을 적습니다.

$y_0 = \frac{exp(x_0)}{u}, y_1 = \frac{exp(x_1)}{u}, y_2 = \frac{exp(x_2)}{u}$


```python
def softmax(x0, x1, x2):
  u = np.exp(x0) + np.exp(x1) + np.exp(x2)
  return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u

y = softmax(2, 1, -1)
print(np.round(y, 2))
print(np.sum(y))
```

```python
from mpl_toolkits.mplot3d import Axes3D

xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)

y = np.zeros((xn, xn, 3))
for i0 in range(xn):
  for i1 in range(xn):
    y[i1, i0, :] = softmax(x0[i0], x1[i1], 1) #x2 = 1로 고정하여 그리기

xx0, xx1 = np.meshgrid(x0, x1) # 두 좌표로 가능한 모든 조합을 만들어내기
plt.figure(figsize=(8, 3))
for i in range(2):
  ax = plt.subplot(1, 2, i+1, projection='3d')
  ax.plot_surface(xx0, xx1, y[:, :, i], rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
  ax.set_xlabel('$x_0$', fontsize=14)
  ax.set_ylabel('$x_1$', fontsize=14)
  ax.view_init(40, -125)

plt.show()
```

### 4.7.7 소프트맥스 함수와 시그모이드 함수

2변수의 경우 소프트맥스 함수

$y = \frac{e^{x_0}}{e^{x_0}+e^{x_1}}$

여기서 $x = x_0 - x_1$로 두면 시그모이드 함수가 된다.

$y = \frac{1}{1+e^{-x}}$

즉 두 변수의 소프트맥스 함수의 입력 $x_0, x_1$ 을 그 차이 $x = x_0 - x_1$로 나타낸 것이 시그모이드 함수이다. 시그모이드 함수를 다변수로 확장한 것이 소프트맥스 함수이다.

### 2차원 가우스 함수

입력이 2차원 벡터 $x = \{x_0, x_1\}^T$라고 했을 때 가우스 함수의 기본형

$y = exp\{-(x_0^2+x_1^2)\}$ 이며 원점을 중심으로 동심원을 가진 종 모양이 된다.

image.png


기본형으로 중심을 이동시키거나 길고 가늘게 만들기 위해 몇 가지 매개 변수를 더한 형태

image.png

함수의 형태를 나타내는 매개변수는 $μ$와 $Σ$입니다. $μ$는 평균 벡터(중심 벡터)로 불리는 매개 변수이며, 분포의 중심을 나타냅니다.

$μ= \begin{bmatrix} μ_0&μ_1\\ \end{bmatrix}^T$

$Σ$는 공분산 행렬로 2X2 행렬입니다

$Σ = \begin{bmatrix} cov(x, x)&cov(x, y)\\cov(y, x)&cov(y, y)\\ \end{bmatrix}$


$Σ=\begin{bmatrix} σ_0&σ_2\\σ_2&σ_1\\ \end{bmatrix}$

이 때 $σ_0$는 $x_0$, $σ_1$는 $x_1$방향으로 얼마나 퍼져있는지,  $σ_2$는 분포의 기울기에 대응하며, 양수 : 오른쪽으로 치우침, 음수 : 왼쪽으로 치우침 분포가 표현됩니다.

exp는 벡터나 행렬이 들어가 있지만 정리하면 스칼라 양이라고 볼 수 있다. 단순하게 $μ= \begin{bmatrix} μ_0&μ_1\\ \end{bmatrix}^T = \begin{bmatrix}0&0\\ \end{bmatrix}^T$라고 했을 때 exp는 $x_0$과 $x_1$의 2차식임을 알 수 있다.

image.png

a는 함수의 크기를 제어하는 매개 변수로도 생각할 수 있지만 가우스 함수에서 확률 분포를 나타내는 경우에는 이와 같이 설정함

image.png

이러면 입력 공간에서의 적분 값이 1이고, 함수가 확률 분포를 표현할 수 있다. 이 때 $|Σ|$는 $Σ$의 행렬식이다. 이 때 $|Σ|$는 이런 식으로 나타낼 수 있다.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def gauss(x, mu, sigma):
  N, D = x.shape
  c1 = 1 / (2*np.pi) ** (D / 2)
  c2 = 1 / (np.linalg.det(sigma) ** (1 / 2))
  inv_sigma = np.linalg.inv(sigma)
  c3 = x - mu
  c4 = np.dot(c3, inv_sigma)
  c5 = np.zeros(N)
  for d in range(D):
    c5 = c5 + c4[:, d] * c3[:, d]
  p = c1 * c2 * np.exp(-c5 / 2)
  return p
```

```python
X_range0=[-3,3]
X_range1=[-3,3]

def show_contour_gauss(mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')] #order = 'F'를 통해 열 우선 정렬
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    cont = plt.contour(xx0, xx1, f, 15, colors='k')
    plt.grid(True)

def show3d_gauss(ax, mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, xn * xn, 'F'), np.reshape(xx1, xn * xn, 'F')]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    ax.plot_surface(xx0, xx1, f, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')


mu = np.array([1, 0.5])
sigma = np.array([[2, 1], [1, 1]])
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
